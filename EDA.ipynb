```python

data['original_price'] = data['original_price'].str.replace('$',' ', regex=True )
data['original_price'] = data['original_price'].astype(float)
```

```python

data = data.fillna( method = 'pad', axis = 1)
```

```python
data.dtypes
```

```python
data = data.drop(['brand', 'crawled_at', 'country', 'language', 'currency','source','source_website','url', 'images','sku','description'], axis=1)
```
